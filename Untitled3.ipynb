```python
# ============================================================
# Corporate Bankruptcy Prediction
# Logistic Regression vs Random Forest vs Decision Tree
# Includes High-Quality Decision Tree Visualization (dtreeviz)
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import (
    accuracy_score,
    recall_score,
    confusion_matrix,
    roc_curve,
    roc_auc_score
)

sns.set(style="whitegrid")

# ============================================================
# 1. Load and preprocess data
# ============================================================

df = pd.read_csv("data.csv")

# Strip whitespace from column names
df.columns = df.columns.str.strip()

if "Bankrupt?" not in df.columns:
    raise ValueError("Target column 'Bankrupt?' not found. Check your CSV.")

X = df.drop(columns=["Bankrupt?"])
y = df["Bankrupt?"]

print("Dataset shape:", df.shape)
display(df.head())

# ============================================================
# 2. Train-test split + scaling
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

feature_names = X.columns.tolist()

print("Training samples:", len(X_train))
print("Test samples:", len(X_test))

# ============================================================
# 3. Fit models
# ============================================================

logit = LogisticRegression(
    class_weight="balanced",
    max_iter=1000,
    solver="lbfgs",
    random_state=42
)
logit.fit(X_train_scaled, y_train)

rf = RandomForestClassifier(
    class_weight="balanced",
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train_scaled, y_train)

dt = DecisionTreeClassifier(
    class_weight="balanced",
    max_depth=5,  # good depth for visualization
    random_state=42
)
dt.fit(X_train_scaled, y_train)

# ============================================================
# 4. Logistic Regression: Top 10 coefficients
# ============================================================

coefs = logit.coef_.flatten()
coef_df = pd.DataFrame({
    "Feature": feature_names,
    "Coefficient": coefs,
    "AbsCoef": np.abs(coefs)
})

top10 = coef_df.sort_values("AbsCoef", ascending=False).head(10)
print("\nTop 10 most significant features (Logit):")
display(top10[["Feature", "Coefficient"]])

# ============================================================
# 5. Evaluation function
# ============================================================

def evaluate(model, X_test, y_test, name):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    sensitivity = recall_score(y_test, y_pred, pos_label=1)
    specificity = recall_score(y_test, y_pred, pos_label=0)
    y_proba = model.predict_proba(X_test)[:, 1]
    return {
        "Model": name,
        "Accuracy": accuracy,
        "Sensitivity": sensitivity,
        "Specificity": specificity,
        "y_pred": y_pred,
        "y_proba": y_proba
    }

logit_res = evaluate(logit, X_test_scaled, y_test, "Logistic Regression")
rf_res = evaluate(rf, X_test_scaled, y_test, "Random Forest")
dt_res = evaluate(dt, X_test_scaled, y_test, "Decision Tree")

# ============================================================
# 6. Comparison table
# ============================================================

comparison = pd.DataFrame([
    {k: v for k, v in logit_res.items() if k not in ["y_pred", "y_proba"]},
    {k: v for k, v in rf_res.items() if k not in ["y_pred", "y_proba"]},
    {k: v for k, v in dt_res.items() if k not in ["y_pred", "y_proba"]},
])

print("\nModel performance comparison:")
display(comparison)

# ============================================================
# 7. Confusion matrices
# ============================================================

def plot_cm(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(4, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

plot_cm(y_test, logit_res["y_pred"], "Logistic Regression - Confusion Matrix")
plot_cm(y_test, rf_res["y_pred"], "Random Forest - Confusion Matrix")
plot_cm(y_test, dt_res["y_pred"], "Decision Tree - Confusion Matrix")

# ============================================================
# 8. ROC curves
# ============================================================

plt.figure(figsize=(7, 6))

for res in [logit_res, rf_res, dt_res]:
    fpr, tpr, _ = roc_curve(y_test, res["y_proba"])
    auc = roc_auc_score(y_test, res["y_proba"])
    plt.plot(fpr, tpr, label=f"{res['Model']} (AUC={auc:.3f})")

plt.plot([0, 1], [0, 1], "k--", label="Random")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()

# ============================================================
# 9. Decision Tree Visualization (dtreeviz)
# ============================================================

# Install dtreeviz if running in Colab:
# !pip install dtreeviz

from dtreeviz.trees import dtreeviz

viz = dtreeviz(
    dt,
    X_train_scaled,
    y_train,
    target_name="Bankrupt?",
    feature_names=feature_names,
    class_names=["Stable", "Bankrupt"]
)

viz

# ============================================================
# 10. Backup visualization using sklearn (always works)
# ============================================================

plt.figure(figsize=(20, 10))
plot_tree(
    dt,
    feature_names=feature_names,
    class_names=["Stable", "Bankrupt"],
    filled=True,
    rounded=True,
    fontsize=8
)
plt.title("Decision Tree (sklearn)")
plt.show()
```
